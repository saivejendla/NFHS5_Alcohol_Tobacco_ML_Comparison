# Outliers Treatment using Capping Approach
   ### IQR Method to treat Outliers:

All the values below Q1 - 1.5*IQR and values above Q3 + 1.5*IQR are outliers and can be removed.


```python
# finding the Quantiles:

Q1 = sub_table_loc.Women_age_15_years_and_above_who_use_any_kind_of_tobacco.quantile(0.25)
Q2 = sub_table_loc.Women_age_15_years_and_above_who_use_any_kind_of_tobacco.quantile(0.50)
Q3 = sub_table_loc.Women_age_15_years_and_above_who_use_any_kind_of_tobacco.quantile(0.75)

# IQR : Inter-Quartile Range

IQR = Q3 - Q1

# Lower Limit:
LC = Q1 - (1.5*IQR)

# Upper Limit:
UC = Q3 + (1.5*IQR)

display(LC)
display(UC)
```

```python
## Plot

sns.distplot(sub_table_loc.Women_age_15_years_and_above_who_use_any_kind_of_tobacco)
plt.axvline(UC, color='r')
plt.axvline(LC, color ='r')
plt.axvline(Q1, color='g')
plt.axvline(Q3, color='g')
plt.show()
```

```python
# Find count of Outliers wrt IQR

sub_table_loc[(sub_table_loc.Women_age_15_years_and_above_who_use_any_kind_of_tobacco<LC) | (sub_table_loc.Women_age_15_years_and_above_who_use_any_kind_of_tobacco>UC)].reset_index(drop=True)
```

```python
sub_table_loc[(sub_table_loc.Women_age_15_years_and_above_who_use_any_kind_of_tobacco<LC) | (sub_table_loc.Women_age_15_years_and_above_who_use_any_kind_of_tobacco>UC)].shape[0]
```

```python
## Store the clean data wrt IQR:

df2 = sub_table_loc[(sub_table_loc.Women_age_15_years_and_above_who_use_any_kind_of_tobacco>LC) & (sub_table_loc.Women_age_15_years_and_above_who_use_any_kind_of_tobacco<UC)]
df2.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Prepare the Data
X = df2[['Women_age_15_years_and_above_who_use_any_kind_of_tobacco', 'Men_age_15_years_and_above_who_use_any_kind_of_tobacco']]
Y = df2['tobacco_consumption_percentage']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

# Initialize models
models = {
    'Linear Regression': LinearRegression(),
    'K-Nearest Neighbors': KNeighborsRegressor(n_neighbors=5),
    'Decision Tree': DecisionTreeRegressor(random_state=0),
    'Random Forest': RandomForestRegressor(random_state=0),
    'Gradient Boosting': GradientBoostingRegressor(random_state=0)
}

# Initialize dictionaries to store metrics
mse_metrics = {}
r2_metrics = {}

# Train and evaluate each model
for name, model in models.items():
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    mse = mean_squared_error(Y_test, Y_pred)
    r2 = r2_score(Y_test, Y_pred)
    
    mse_metrics[name] = mse
    r2_metrics[name] = r2

# Plotting MSE and R-squared metrics
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Plot Mean Squared Error (MSE)
ax[0].bar(mse_metrics.keys(), mse_metrics.values(), color='blue')
ax[0].set_xlabel('Models')
ax[0].set_ylabel('Mean Squared Error (MSE)')
ax[0].set_title('Mean Squared Error for Different Regression Models')
ax[0].set_xticklabels(mse_metrics.keys(), rotation=45, ha='right')

# Adding value labels on top of bars
for i, (model, value) in enumerate(mse_metrics.items()):
    ax[0].text(i, value + 0.02, f'{value:.4f}', ha='center', va='bottom')

# Plot R-squared
ax[1].bar(r2_metrics.keys(), r2_metrics.values(), color='orange')
ax[1].set_xlabel('Models')
ax[1].set_ylabel('R-squared')
ax[1].set_title('R-squared for Different Regression Models')
ax[1].set_xticklabels(r2_metrics.keys(), rotation=45, ha='right')

# Adding value labels on top of bars
for i, (model, value) in enumerate(r2_metrics.items()):
    ax[1].text(i, value + 0.02, f'{value:.4f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
