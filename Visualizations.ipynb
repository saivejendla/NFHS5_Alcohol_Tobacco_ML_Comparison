# Comparison by gender

```python
import pandas as pd
import matplotlib.pyplot as plt

# Import the Healthcare dataset from an Excel file
df = pd.read_excel("Healthcare.xls")

# Columns names
men_col = 'Men age 15 years and above who consume alcohol (%)'
women_col = 'Women age 15 years and above who consume alcohol (%)'

# Filter the dataset for the specific states of interest
states_of_interest = ['Andhra Pradesh', 'Karnataka', 'Uttar Pradesh', 'Madhya Pradesh', 'Tamil Nadu']
filtered_data = df[df['State/UT'].isin(states_of_interest)]

# Create a new dataframe with state names and the two columns of interest
data = filtered_data[['State/UT', men_col, women_col]]

# Set the state names as the index
data.set_index('State/UT', inplace=True)

# Sort the data by state names for better visualization
data.sort_index(inplace=True)

# Plot the data
plt.figure(figsize=(12, 8))

# Create a bar plot for the two columns
data.plot(kind='bar', figsize=(12, 8), width=0.8)

# Adding title and labels
plt.title('Alcohol Consumption Comparison by Gender in Selected Indian States', fontsize=20)
plt.xlabel('States/UTs', fontsize=16)
plt.ylabel('Alcohol Consumption (%)', fontsize=16)
plt.xticks(rotation=0)

# Adding legend
plt.legend(['Men', 'Women'], fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

# Decission Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Import the Healthcare dataset from an Excel file
df = pd.read_excel("Healthcare.xls")

# Assuming that you have a target variable and feature variables
target_column_name = 'Men age 15 years and above who consume alcohol (%)'
feature_columns = ['Women age 15 years and above who consume alcohol (%)']



# Separate the features (X) and the target variable (y)
X = df[feature_columns]
y = df[target_column_name]

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.75)

# Create a decision tree regression model
dt_model = DecisionTreeRegressor()

# Fit the model to the training data
dt_model.fit(X_train, y_train)

# Make predictions on the test data
dt_predictions = dt_model.predict(X_test)

# Evaluate the model
print("Decision Tree Regression")
print("Mean squared error:", mean_squared_error(y_test, dt_predictions))
print("R-squared:", r2_score(y_test, dt_predictions))
```

```python
# Plot the data points and the regression line
plt.scatter(X_test, y_test, color='blue', label='Actual Data')
plt.scatter(X_test, dt_predictions, color='red', label='Predictions')
plt.xlabel('Women who consume alcohol (%)')
plt.ylabel('Men who use alcohol (%)')
plt.title('Decision Tree Regression')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Separate the features (X) and the target variable (y)
X = df[feature_columns]
y = df[target_column_name]

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.75)

# Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_predictions = lr_model.predict(X_test)
lr_mse = mean_squared_error(y_test, lr_predictions)
lr_r2 = r2_score(y_test, lr_predictions)

# Decision Tree Regression
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)
dt_predictions = dt_model.predict(X_test)
dt_mse = mean_squared_error(y_test, dt_predictions)
dt_r2 = r2_score(y_test, dt_predictions)

# Random Forest Regression
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
rf_mse = mean_squared_error(y_test, rf_predictions)
rf_r2 = r2_score(y_test, rf_predictions)

# Print the evaluation metrics
print("Linear Regression: MSE =", lr_mse, ", R-squared =", lr_r2)
print("Decision Tree Regression: MSE =", dt_mse, ", R-squared =", dt_r2)

# Visualization with lighter colors
models = ['Linear Regression', 'Decision Tree', 'Random Forest']
mse_values = [lr_mse, dt_mse, rf_mse]
r2_values = [lr_r2, dt_r2, rf_r2]

x = np.arange(len(models))  # the label locations
width = 0.35  # the width of the bars

fig, ax1 = plt.subplots(figsize=(12, 6))

print("Random Forest Regression: MSE =", rf_mse, ", R-squared =", rf_r2)
ax2 = ax1.twinx()
rects1 = ax1.bar(x - width/2, mse_values, width, label='MSE', color='#99c2ff')  # Light blue for MSE
rects2 = ax2.bar(x + width/2, r2_values, width, label='R-squared', color='#ffb3e6')  # Light pink for R-squared

# Add some text for labels, title and custom x-axis tick labels, etc.
ax1.set_xlabel('Models')
ax1.set_ylabel('Mean Squared Error', color='#99c2ff')  # Matching color for MSE label
ax2.set_ylabel('R-squared', color='#ffb3e6')  # Matching color for R-squared label
ax1.set_title('Comparison of Regression Models')
ax1.set_xticks(x)
ax1.set_xticklabels(models)
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

```python
# Separate the features (X) and the target variable (y)
X = df[feature_columns]
y = df[target_column_name]

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.75)

# Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_predictions = lr_model.predict(X_test)
lr_mse = mean_squared_error(y_test, lr_predictions)
lr_r2 = r2_score(y_test, lr_predictions)

# Decision Tree Regression
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)
dt_predictions = dt_model.predict(X_test)
dt_mse = mean_squared_error(y_test, dt_predictions)
dt_r2 = r2_score(y_test, dt_predictions)

# Random Forest Regression
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
rf_mse = mean_squared_error(y_test, rf_predictions)
rf_r2 = r2_score(y_test, rf_predictions)

# Determine the best model based on MSE and R-squared
models = ['Linear Regression', 'Decision Tree', 'Random Forest']
mse_values = [lr_mse, dt_mse, rf_mse]
r2_values = [lr_r2, dt_r2, rf_r2]

best_model_index_mse = np.argmin(mse_values)
best_model_index_r2 = np.argmax(r2_values)

best_model_name_mse = models[best_model_index_mse]
best_model_name_r2 = models[best_model_index_r2]

# Print the best models based on MSE and R-squared
print(f"Best model based on MSE: {best_model_name_mse} (MSE = {mse_values[best_model_index_mse]})")
print(f"Best model based on R-squared: {best_model_name_r2} (R-squared = {r2_values[best_model_index_r2]})")

# Visualization with lighter colors
x = np.arange(len(models))  # the label locations
width = 0.35  # the width of the bars

fig, ax1 = plt.subplots(figsize=(12, 6))

ax2 = ax1.twinx()
rects1 = ax1.bar(x - width/2, mse_values, width, label='MSE', color='#99c2ff')  # Light blue for MSE
rects2 = ax2.bar(x + width/2, r2_values, width, label='R-squared', color='#ffb3e6')  # Light pink for R-squared

# Highlight the best model based on MSE
rects1[best_model_index_mse].set_color('#1f77b4')  # Darker blue for best MSE

# Highlight the best model based on R-squared
rects2[best_model_index_r2].set_color('#ff7f0e')  # Darker orange for best R-squared

# Add some text for labels, title and custom x-axis tick labels, etc.
ax1.set_xlabel('Models')
ax1.set_ylabel('Mean Squared Error', color='#1f77b4')  # Matching color for MSE label
ax2.set_ylabel('R-squared', color='#ff7f0e')  # Matching color for R-squared label
ax1.set_title('Comparison of Regression Models')
ax1.set_xticks(x)
ax1.set_xticklabels(models)
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
